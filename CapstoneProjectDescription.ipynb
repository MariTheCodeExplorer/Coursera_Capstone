# "Where should I eat" recommender system

## Introduction/Business Problem 

Recommender system will offer users "Best match" of dining places. 
Places where user has visited will be analysed, what type of food or drinks do they offer, their price rate, popularity and 
other relevant attributes will be used for analysis. Location will be taken into consideration for make match by closest "Best match" place.
Time of day will be as well taken into consideration, to know if the place is open when user is using recommender system and 
whether or not it is the time of day when user visits similar places(For example if user usually visits coffee shops in the morning, bar will not be the "best match").
In overgrowing cities, where people get numerous choices  of somewhat similar places, making decision of where to eat has become rather time consuming. For indecisive people  who may end up overly analyzing their choices of where to eat and spending their free time on deciding where to go instead of  using  it more appropriately, recommender system that will do the analysis for them will be time and nerve saviour .

## Data

For data source of location data I will use Foursquare Check-in data for Tallinn city (more in detail the fields and their description can be found here: https://developer.foursquare.com/docs/api/checkins/details) and I will use User-User collaborative filtering algorithms for recommender system.
